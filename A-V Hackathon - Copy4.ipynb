# Preprocessing the data

```python
train.isnull().sum()
```

```python
train['application_underwriting_score'].fillna(train['application_underwriting_score'].mean(),inplace=True)
```

```python
#converting sourcing_channel into numerical columns
sourcing_channel=pd.get_dummies(train['sourcing_channel'],prefix='Channel')
train=pd.concat([train,sourcing_channel],axis=1)
train.drop('sourcing_channel',axis=1,inplace=True)
```

```python
#converting residence_area_type into numerical columns
residence_area_type=pd.get_dummies(train['residence_area_type'],prefix='area')
train=pd.concat([train,residence_area_type],axis=1)
train.drop('residence_area_type',axis=1,inplace=True)
```

```python
test.isnull().sum()
```

```python
test['application_underwriting_score'].fillna(test['application_underwriting_score'].mean(),inplace=True)
```

```python
#converting sourcing_channel into numerical columns
sourcing_channel=pd.get_dummies(test['sourcing_channel'],prefix='Channel')
test=pd.concat([test,sourcing_channel],axis=1)
test.drop('sourcing_channel',axis=1,inplace=True)
```

```python
##converting residence_area_type into numerical columns
residence_area_type=pd.get_dummies(test['residence_area_type'],prefix='area')
test=pd.concat([test,residence_area_type],axis=1)
test.drop('residence_area_type',axis=1,inplace=True)
```

```python
train.fillna(0,inplace=True)
test.fillna(0,inplace=True)
```

# Feature selection

```python
feature_cols=['age_in_days','perc_premium_paid_by_cash_credit','Income','Count_3-6_months_late','Count_6-12_months_late', 'Count_more_than_12_months_late','no_of_premiums_paid','application_underwriting_score','Channel_A','Channel_B','Channel_C','Channel_D','area_Urban','area_Rural']
X=train[feature_cols]
y=train['target']
```

# Splitting Data

```python
from sklearn.model_selection import train_test_split
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
#importing decision tree classifier 
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
# Model Accuracy, how often is the classifier correct?
accuracy_score(y_test, y_pred)
```

```python
test_cols=['age_in_days','perc_premium_paid_by_cash_credit','Income','Count_3-6_months_late','Count_6-12_months_late', 'Count_more_than_12_months_late','no_of_premiums_paid','application_underwriting_score','Channel_A','Channel_B','Channel_C','Channel_D','area_Urban','area_Rural']
test_updated =test[test_cols]
```

```python
predicted = clf.predict(test_updated)
```

```python
submission =pd.DataFrame(predicted)
submission.index = test['id']
submission.columns = ['target']
```
